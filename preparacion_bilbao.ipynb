```python
import requests
import pandas as pd
import os
import re
```

```python
df_bilbao = pd.read_csv("bilbao_csv\datosbilbao.csv", sep =",")
df_bilbao
```

```python
#Separamos la columna "fecha" con año-mes en dos
df_bilbao[['año', 'mes']] = df_bilbao['fecha'].str.split('-', expand=True)
# Eliminamos la columna "fecha"
df_bilbao2 = df_bilbao.drop(columns=["fecha"], axis=1)
# Ponemos las nuevas columnas año, mes las primeras
column_order = ['año', 'mes'] + df_bilbao2.columns.tolist()[:-2]
df_bilbao3 = df_bilbao2.reindex(columns=column_order)
df_bilbao3
df_bilbao3['ta_max'] = df_bilbao3['ta_max'].str.replace(r'\(.*\)', '', regex=True)
df_bilbao3['ta_min'] = df_bilbao3['ta_min'].str.replace(r'\(.*\)', '', regex=True)
df_bilbao3
# Así se pasan las filas de los datos anuales a otro dataframe
df_bilbao_anual = pd.DataFrame(df_bilbao3.iloc[12::13])
df_bilbao_anual
# Y eliminamos esas filas del dataframe original
df_bilbao_final = df_bilbao3.drop(range(12,710,13), axis=0)
df_bilbao_final
#Generamos csv y xlsx para guardar los archivos y que no se pierda el trabajo
df_bilbao_anual.to_csv('bilbao_anual.csv', index=False)
df_bilbao_final.to_csv('bilbao_final.csv', index=False)
df_bilbao_anual.to_excel('bilbao_anual.xlsx', index=False)
df_bilbao_final.to_excel('bilbao_final.xlsx', index=False)
```

```python
#Después de eliminar las filas nº 13, hace falta resetear el indice. drop=True indica que elimine el índice antiguo, inplace=True que lo modifique en el mismo dataframe sin crear otro
df_bilbao_final.reset_index(drop=True, inplace=True)
```
