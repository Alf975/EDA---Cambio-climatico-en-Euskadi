```python
import requests
import pandas as pd
```

```python
direccion="https://opendata.aemet.es/opendata/sh/c834d5a3"

def hacer_dataframe(info):
        response = requests.get(info)
        data = response.json()
        df_inicial = pd.DataFrame(data)
        df_filtrado = pd.DataFrame(df_inicial[["fecha", "tm_mes", "tm_max", "tm_min", "p_mes"]])
        df_12filas = df_filtrado.drop([12], axis=0)
        # datos_finales = pd.concat([datos_finales, df_12filas], axis=0)
        return df_12filas

df_1970 = pd.DataFrame(hacer_dataframe(direccion))

df_1970.to_csv('datos1970.csv', index=False)
df_1970.to_excel('datos1970.xlsx', index=False)
```

```python
#Funcion para importar csv y concatenar dataframes.
import os

df_concatenado = []
def importar_concatenar(carpeta):
    for archivo in os.listdir(carpeta):
        ruta_archivo = os.path.join(carpeta, archivo)
        df = pd.read_csv(ruta_archivo, sep = ',')
        df_concatenado = pd.concat([df_concatenado,df])
    return df_concatenado



df_finalizado = pd.DataFrame(importar_concatenar("C:\Users\adelm\OneDrive\Bootcamp\GitHub mio\Core\1-Data_Analysis\Entregas\EDA\archivos_csv"))        
```
